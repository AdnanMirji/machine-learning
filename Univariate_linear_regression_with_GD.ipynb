```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data=pd.read_csv('/content/drive/MyDrive/Task_1/univar.csv')
print(data)

x=np.array(data[['x']])
y=np.array(data[['y']])

print(x)
print(y)

plt.scatter(x,y)
plt.xlabel('independent variable')
plt.ylabel('dependent variable')
plt.show()
threshold=0.0001
alpha=0.01
iterations=1000
m=0
b=0

for i in range(iterations):
  y_pred=m*x+b
  error=y_pred-y

  m=m-alpha*(2/len(y))* np.sum(error*x)
  b=b-alpha*(2/len(y))*np.sum(error)
  cost=(1/(2*len(y))) *np.sum((error)**2)
  if  cost < threshold:
    break


print("optimized m and b",m,b)

y_predict=m*x+b
plt.scatter(x,y)
plt.plot(x,y_predict)
plt.xlabel('independent variable')
plt.ylabel('dependent variable')
plt.show()

```
