```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv('/content/drive/MyDrive/Task_1/logregress.csv.csv')
print(data)

x=np.array(data[['Exam1','Exam2']])
y=np.array(data[['addmission']])

print(x)
print(y)

x_b=np.c_[np.ones((len(x),1)),x]
print(x_b)
weight=np.zeros((x_b.shape[1],1))
print(weight)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

alpha=0.01
iterations=1000

for i in range(iterations):
  z=x_b.dot(weight)
  y_prob=sigmoid(z)
  error=y_prob-y
  gradient=x_b.T.dot(error)
  weight=weight-(alpha*gradient)

print(weight)

plt.scatter(x[:,0],x[:,1], c=y.ravel(), cmap=plt.cm.Paired, edgecolors='k', marker='o')
plt.xlabel('Exam1')
plt.ylabel('Exam2')

x1_boundary=np.array([min(x[:,0]),max(x[:,0])])
print(x1_boundary)
x2_boundary=-(weight[0]+weight[1]*x1_boundary)/weight[2]
print(x2_boundary)

plt.plot(x1_boundary,x2_boundary,color='red',linestyle='--',label='decision boundary')
plt.legend
plt.show
```
